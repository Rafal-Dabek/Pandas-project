```python
import pandas as pd
import matplotlib.pyplot as plt
pilkarze = pd.read_csv('players_21.csv', sep=',')

del pilkarze['sofifa_id']
del pilkarze['player_url']
del pilkarze['long_name']
del pilkarze['dob']
del pilkarze['rb']
del pilkarze['rcb']
del pilkarze['cb']
del pilkarze['lcb']
del pilkarze['lb']
del pilkarze['rwb']
del pilkarze['rdm']
del pilkarze['cdm']
del pilkarze['ldm']
del pilkarze['lwb']
del pilkarze['rm']
del pilkarze['rcm']
del pilkarze['cm']
del pilkarze['lcm']
del pilkarze['lm']
del pilkarze['ram']
del pilkarze['cam']
del pilkarze['lam']
del pilkarze['rw']
del pilkarze['rf']
del pilkarze['cf']
del pilkarze['lf']
del pilkarze['lw']
del pilkarze['rs']
del pilkarze['ls']
del pilkarze['st']
del pilkarze['defending_marking']
pilkarze['quality']=pilkarze['value_eur']/pilkarze['overall']


del pilkarze['league_rank']
del pilkarze['real_face']
del pilkarze['loaned_from']




manu=pilkarze.loc[pilkarze['club_name']=="Manchester United"].sort_values('wage_eur',ascending=False)

bins=[60,65,70,75,80,85,90]

plt.hist(manu.overall,bins=bins)

plt.xticks(bins)

plt.show()



bpl_players=pilkarze.loc[pilkarze['league_name']=="English Premier League"].sort_values('wage_eur',ascending=False)
dane_bpl=bpl_players["value_eur"][0:10]

pilkarze.describe()
```

```python
psg=pilkarze.loc[pilkarze['club_name']=="Paris Saint-Germain"].sort_values('wage_eur',ascending=False)

dane15=psg["overall"][0:10]

plt.plot(range(len(dane15)),dane15,"ro",label="Overalle 10 najlepiej zarabiających piłkarzy PSG")
```

```python
print("Średnia wieku piłkarzy w Manchester United i piłkarze młodsi niż średnia")
manu=pilkarze.loc[pilkarze['club_name']=="Manchester United"]
manu.age.mean()
younger_manu=manu.loc[pilkarze['age']<manu.age.mean()]
younger_manu
```

```python
print("Gracze mający overall dla Premier League, Bundesligi,La Ligi i Serie A większy od 85")
maxovrPL=pilkarze.loc[(pilkarze['league_name']=="English Premier League") & (pilkarze['overall']>int(85))]
maxovrBund=pilkarze.loc[(pilkarze['league_name']=="German 1. Bundesliga") & (pilkarze['overall']>int(85))]
maxovrLaliga=pilkarze.loc[(pilkarze['league_name']=="Spain Primera Division") & (pilkarze['overall']>int(85))]
maxovrSerieA=pilkarze.loc[(pilkarze['league_name']=="Italian Serie A") & (pilkarze['overall']>int(85))]

maxovrPL
maxovrBund
maxovrLaliga
maxovrSerieA
```

```python
lewa_noga=pilkarze.loc[(pilkarze['preferred_foot']=='Left') & (pilkarze['nationality']=='Poland')].count()[0]
prawa_noga=pilkarze.loc[(pilkarze['preferred_foot']=='Right') & (pilkarze['nationality']=='Poland')].count()[0]

labels=['Lewa noga','Prawa noga']

plt.pie([lewa_noga,prawa_noga],labels=labels,autopct='%.2f %%')
plt.title("Wykres procentowy prawo i lewonożnych polskich piłkarzy")
plt.show()
```

```python


polPL=pilkarze.loc[(pilkarze['league_name']=='English Premier League') & (pilkarze['nationality']=='Poland')].count()[0]
polBund=pilkarze.loc[(pilkarze['league_name']=='German 1. Bundesliga') & (pilkarze['nationality']=='Poland')].count()[0]
polLig1=pilkarze.loc[(pilkarze['league_name']=='French Ligue 1') & (pilkarze['nationality']=='Poland')].count()[0]
polLalig=pilkarze.loc[(pilkarze['league_name']=='Spain Primera Division') & (pilkarze['nationality']=='Poland')].count()[0]
polSerieA=pilkarze.loc[(pilkarze['league_name']=='Italian Serie A') & (pilkarze['nationality']=='Poland')].count()[0]



labels=['Premier League','Bundesliga','La Liga','Serie A','Ligue 1']

plt.pie([polPL,polBund,polLalig,polSerieA,polLig1],labels=labels,autopct='%.2f %%')
plt.title("Wykres procentowy lig polskich piłkarzy")
plt.show()
```
