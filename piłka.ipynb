```python
import pandas as pd
import matplotlib.pyplot as plt
pilkarze = pd.read_csv('players_21.csv', sep=',')

del pilkarze['sofifa_id']
del pilkarze['player_url']
del pilkarze['long_name']
del pilkarze['dob']
del pilkarze['rb']
del pilkarze['rcb']
del pilkarze['cb']
del pilkarze['lcb']
del pilkarze['lb']
del pilkarze['rwb']
del pilkarze['rdm']
del pilkarze['cdm']
del pilkarze['ldm']
del pilkarze['lwb']
del pilkarze['rm']
del pilkarze['rcm']
del pilkarze['cm']
del pilkarze['lcm']
del pilkarze['lm']
del pilkarze['ram']
del pilkarze['cam']
del pilkarze['lam']
del pilkarze['rw']
del pilkarze['rf']
del pilkarze['cf']
del pilkarze['lf']
del pilkarze['lw']
del pilkarze['rs']
del pilkarze['ls']
del pilkarze['st']
del pilkarze['defending_marking']
pilkarze['value/overall']=pilkarze['value_eur']/pilkarze['overall']
#pilkarze.describe()
#pilkarze.sort_values('wage_eur',ascending='False')
del pilkarze['league_rank']
del pilkarze['real_face']
del pilkarze['loaned_from']


manu=pilkarze.loc[pilkarze['club_name']=="Manchester United"].sort_values('wage_eur',ascending=False)

dane14=manu["overall"][0:10]

#plt.plot(range(len(dane14)),dane14,"ro",label="2014")

psg=pilkarze.loc[pilkarze['club_name']=="Paris Saint-Germain"].sort_values('wage_eur',ascending=False)

dane15=psg["overall"][0:10]

#plt.plot(range(len(dane15)),dane15,"ro",label="2014")

bins=[60,65,70,75,80,85,90]

plt.hist(manu.overall,bins=bins)

plt.xticks(bins)

plt.show()

#pilkarze.sort_values('value/overall',ascending=False)

bpl_players=pilkarze.loc[pilkarze['league_name']=="English Premier League"].sort_values('wage_eur',ascending=False)
dane_bpl=bpl_players["value_eur"][0:10]
dane_bpl
bpl_players

```

```python
print("Najstarszy zawodnik: " )
pilkarze[pilkarze.age==pilkarze.age.max()]
```

```python
print("Najmłodsi zawodnicy:")
pilkarze[pilkarze.age==pilkarze.age.min()]
```

```python
print("Najstarsi/najmłodsi gracze z Polski:")
pilkarze.loc[pilkarze['nationality']=="Poland"].sort_values('age',ascending=False)
```

```python
wiekpolska=pilkarze.loc[pilkarze['nationality']=="Poland"]

bins=[16,20,25,30,35,40]

plt.hist(wiekpolska.age, bins=bins)

plt.xticks(bins)

plt.ylabel("Polscy piłkarze")
plt.xlabel("Wiek piłkarzy")
plt.title("Wykres kołowy przedstawiający rozkład wieku polskich piłkarzy")
plt.show()
```

```python
wiekniemcy=pilkarze.loc[pilkarze['nationality']=="Germany"]
bins=[16,20,25,30,35,40]

plt.hist(wiekniemcy.age, bins=bins)

plt.xticks(bins)

plt.ylabel("Niemieccy piłkarze")
plt.xlabel("Wiek piłkarzy")
plt.title("Wykres kołowy przedstawiający rozkład wieku niemieckich piłkarzy")
plt.show()
```

```python
print("Piłkarze warci ponad 50 milionów euro:")
#manu=pilkarze.loc[pilkarze['club_name']=="Manchester United"].sort_values('wage_eur',ascending=False)
wartosc1=pilkarze.loc[pilkarze['value_eur']>int(50000000)]
wartosc1
```

```python
print("Piłkarze warci mniej niż 5000 euro")
wartosc2=pilkarze.loc[pilkarze['value_eur']<int(5000)]
wartosc2
```

```python
print("Polscy piłkarze warci więcej niż 1 milion euro:")
polpilk=pilkarze.loc[(pilkarze['nationality']=="Poland") & (pilkarze['value_eur']>int(5000000))]
polpilk

```

```python
print("Niemieccy piłkarze warci więcej niż 1 milion euro:")
niemiecpilk=pilkarze.loc[(pilkarze['nationality']=="Germany") & (pilkarze['value_eur']>int(5000000))]
niemiecpilk
```

```python
print("Gracze w lidze Premier League mający ponad 1.95 wzrostu")
wzrostpl=pilkarze.loc[(pilkarze['league_name']=="English Premier League") & (pilkarze['height_cm']>int(195))]
wzrostpl
```

```python
print("Gracze mający overall dla Premier League, Bundesligi,La Ligi i Serie A większy od 85")
maxovrPL=pilkarze.loc[(pilkarze['league_name']=="English Premier League") & (pilkarze['overall']>int(85))]
maxovrBund=pilkarze.loc[(pilkarze['league_name']=="German 1. Bundesliga") & (pilkarze['overall']>int(85))]
maxovrLaliga=pilkarze.loc[(pilkarze['league_name']=="Spain Primera Division") & (pilkarze['overall']>int(85))]
maxovrSerieA=pilkarze.loc[(pilkarze['league_name']=="Italian Serie A") & (pilkarze['overall']>int(85))]
maxovrPL
maxovrBund
maxovrLaliga
maxovrSerieA
```

```python
lewa_noga=pilkarze.loc[(pilkarze['preferred_foot']=='Left') & (pilkarze['nationality']=='Poland')].count()[0]
prawa_noga=pilkarze.loc[(pilkarze['preferred_foot']=='Right') & (pilkarze['nationality']=='Poland')].count()[0]

labels=['Lewa noga','Prawa noga']

plt.pie([lewa_noga,prawa_noga],labels=labels,autopct='%.2f %%')
plt.title("Wykres procentowy Prawo i Lewonożnych polskich piłkarzy")
plt.show()
```
